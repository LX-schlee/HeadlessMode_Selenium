```python
# import relevant libraries
from selenium import webdriver
import time 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
```

```python
# store website inside a variable
website = 'https://sfbay.craigslist.org/'
```

```python
# initialize Chrome Driver
driver = webdriver.Chrome('C:/webdrivers/chromedriver.exe')

# open website
driver.get(website)

# maximize window
driver.maximize_window()

# variable for search area
search = driver.find_element_by_id('query')

# click on Enter
search.click()
time.sleep(1)

# search for "table"
search.send_keys('table')

# click on Enter
search.send_keys(Keys.ENTER)

# store results (table names) in a variale
tables = driver.find_elements_by_xpath('//h3[@class="result-heading"]')

# print out results
for table in tables:
    print(table.text)
    
# close driver
driver.close()
```

### Headless Mode

```python
chrome_options = Options()
chrome_options.add_argument('--headless')

# initialize Chrome Driver
driver = webdriver.Chrome('C:/webdrivers/chromedriver.exe', options=chrome_options)

# open website
driver.get(website)

print('Website was opened')

# maximize window
driver.maximize_window()

# variable for search area
search = driver.find_element_by_id('query')

# click on Search
search.click()
time.sleep(1)

# search for "table"
search.send_keys('table')

print('We search for tables')

# click on Enter
search.send_keys(Keys.ENTER)


# store results (table names) in a variale
tables = driver.find_elements_by_xpath('//h3[@class="result-heading"]')

print('This are the results: \n')

# print out results
for table in tables:
    print(table.text)
    
# close driver
driver.close()

print('\n Website was closed')
```
